```python
birthdays = {'Leanne': {'month': 'May', 'day': 30},
             'Charles': {'month': 'Oct', 'day': 31},
             'Kuba': {'month': 'May', 'day': 31},
             'Juraj': {'month': 'Feb', 'day': 14},
             'Darren': {'month': 'Apr', 'day': 20}}

months = {'Jan': ['Feb',31], 'Feb': ['Mar',28], 'Mar': ['Apr',31], 'Apr': ['May',30], 
          'May': ['June',31], 'June': ['July',30], 'July': ['Aug',31], 'Aug': ['Sep',31], 
          'Sep': ['Oct',30], 'Oct': ['Nov',31], 'Nov': ['Dec',30], 'Dec': ['Jan',31]}

def printName(name):
    print(name + "'s birthday is on " + birthdays[name]['month'] + " " + str(birthdays[name]['day']))
        
def printMonth(month):
    print("A list of all birthdays on " + month + ":")
    for key, val in birthdays.items():
        if val['month'] == month:
            print(key + ": " + val['month'] + " " + str(val['day']))
            
def printWeek(month1, day1):
    day2 = day1 + 7
    month2 = month1
    max_days = months[month1][1]
    if day2 > max_days:
        day2 = day2 % max_days
        month2 = months[month1][0]
    print("A list of all birthdays between " + month1 + " " + str(day1) + " and " + month2 + " " + str(day2) + ":")
    
    for key, val in birthdays.items():
        birth_month = val['month']
        birth_day = val['day']
        if month1 == month2 and month1 == birth_month and birth_day >= day1 and birth_day <= day2:
            print(key + ": " + birth_month + " " + str(birth_day))
        elif month1 != month2 and ((month1 == birth_month and birth_day >= day1) or (month2 == birth_month and birth_day <= day2)):
            print(key + ": " + birth_month + " " + str(birth_day))
                
    
#printName('Charles')
#printMonth('Feb')
printWeek('May', 29)
```

```python
import pprint as pp

auto_birthdays = {}
def getBirthdays(filename, book):
    with open(filename, 'r') as f:
        for i in f.readlines():
            data = i.strip().split(',')
            book[data[0]] = {'month': data[1], 'day': data[2]}
    return book

pp.pprint(getBirthdays('birthdays.txt', auto_birthdays))
```

```python
import pprint as pp

quit = False
while not quit:
    print('Your wish is my command:')
    print('(Possible wishes: read, search, add, (back, quit))')
    inp = input()
    if inp == 'quit':
        quit = True
        print('Quitting... :(')
    elif inp == 'read':
        accessed = False
        while not accessed and not quit:
            print('    What file would you like to access?')
            filename = input()
            if filename == 'quit':
                quit = True
                print('Quitting... :(')
            elif filename == 'back':
                accessed = True
            else:
                try:
                    dic = getBirthdays(filename, auto_birthdays)
                    print('Reading...')
                    print('Here\'s your list of birthdays:')
                    pp.pprint(dic)
                    accessed = True
                except IOError:
                    print("Sorry, that filename doesn't exist.")
    elif inp == 'search':
        found = False
        while not found and not quit:
            print('    What would you like to search by? (name / month / week)')
            cmd = input()
            if cmd == 'quit':
                quit = True
                print('Quitting... :(')
            elif cmd == 'name':
                print('        What is the name you would like to find?')
                name = input()
                if name == 'quit':
                    quit = True
                    print('Quitting... :(')
                elif name == 'back':
                    found = True
                elif name in birthdays:
                    printName(name)
                    found = True
                else:
                    print("        Sorry, " + name + " not found.")
            elif cmd == 'month':
                print('        What month do you want to filter by?')
                month = input()
                if month == 'quit':
                    quit = True
                    print('Quitting... :(')
                elif month == 'back':
                    found = True
                elif month in months:
                    printMonth(month)
                    found = True
                else:
                    print('Please enter a valid (shortened) month.')
            elif cmd == 'week':
                print('        What month does the week start on?')
                month = input()
                if month == 'quit':
                    quit = True
                    found = True
                    print('Quitting... :(')
                elif month in months:
                    print('        What day of the month does the week start on?')
                    day = input()
                    if day and int(day) > 0 and int(day) <= months[month][1]:
                        printWeek(month, int(day))
                    else:
                        print('        Sorry, that is not a valid day for ' + month)
                else:
                    print('        Sorry, that is not a valid month')
            elif cmd == 'back':
                found = True
            else:
                print('    Sorry, command not recognised.')
    elif inp == 'add':
        added = False
        while not added and not quit:
            print("    What's the name of the birthday person you want to add?")
            name = input()
            if name == 'quit':
                quit = True
                print('Quitting... :(')
            elif name == 'back':
                added = True
            else:
                print("    What's the month of the birthday?")
                month = input()
                if month == 'quit':
                    quit = True
                    print('Quitting... :(')
                elif month == 'back':
                    added = True
                elif month in months:
                    print("    What's the day of the birthday?")
                    day = input()
                    if day == 'quit':
                        quit = True
                        print('Quitting... :(')
                    elif day == 'back':
                        added = True
                    elif day and int(day) > 0 and int(day) <= months[month][1]:
                        birthdays[name] = {'month': month, 'day': day}
                        print('    Entry added!')
                        added = True
                    else:
                        print('    Sorry, that is not a valid day for ' + month)
                else:
                    print("    Sorry, that is not a valid (shortened) month.")
    elif inp == 'list':
        pp.pprint(birthdays)
    else:
        print('Sorry, command not recognised.')
```
